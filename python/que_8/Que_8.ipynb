```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data= pd.read_excel('episodes.xlsx',parse_dates=['airdate','airstamp'])
data
```

```python
# checking datatype and null values
data.info()
```

## Get all the overall ratings for each season and using plots compare the ratings for all the seasons, like season 1 ratings, season 2, and so on.

```python
# lets create avg_rating column as rating column is of type string
avg_rating=[]
for i in data.rating:
    try:
        avg_rating.append(float(i[-4:-1]))
    except:
        avg_rating.append(float(i[-3:-1]))
data['avg_rating']=avg_rating
```

```python
# using group by function we can calculate overall avg rating of a season
data.groupby('season').mean().avg_rating
```

```python
# bar chart did not show significant variation thats why line chart in=s used here.
plt.figure(figsize=(4,3))
data.groupby('season').mean().avg_rating.plot(kind='line')
```

#### from above we can see that rating has drop continuously for each successive season

## Get all the episode names, whose average rating is more than 8 for every season

```python
print(list(data.name[data.avg_rating>8]))
```

## Get all the episode names that aired before May 2019

```python
data.name[[i.year<2019 for i in data.airdate]]
```

## Get the episode name from each season with the highest and lowest rating

```python
min_rating=list(data.groupby('season').min(avg_rating).id)
max_rating=list(data.groupby('season').max(avg_rating).id)
```

```python
print('episode name from each season with the lowest rating'
     ,'\n','season    name')
for i in range(len(data)):
    if data.iloc[i].id in min_rating:
        print(f'   {data.iloc[i,3]}    {data.iloc[i,2]}')
```

```python
print('episode name from each season with the highest'
     ,'\n','season    name')
for i in range(len(data)):
    if data.iloc[i].id in max_rating:
        print(f'   {data.iloc[i,3]}    {data.iloc[i,2]}')
```
