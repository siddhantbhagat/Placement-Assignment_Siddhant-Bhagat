## Get all Pokemons whose spawn time is less than 300 seconds.

```python
data.spawn_time.fillna('00:00',inplace=True)
data.spawn_time
```

```python
spawn_time_sec=[]
for i in data.spawn_time:
    x=int(i[:2])*60+int(i[3:])
    spawn_time_sec.append(x)

data['spawn_time_sec']=spawn_time_sec
```

```python
print(list(data.name[data.spawn_time_sec<300]),end=',')
print('\n',len(list(data.name[data.spawn_time_sec<300])))
```

#### From above we can conclude that there are 81 pokemons who has less than 300 sec span time
