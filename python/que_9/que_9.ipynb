## Get all the cars and their types that do not qualify for clean alternative fuel vehicle

```python
data['Clean Alternative Fuel Vehicle (CAFV) Eligibility'].unique()
```

```python
# as 222 models are missing we have to handel the mission data
data.Model.fillna('Missing',inplace=True)
data.Model.isnull().sum()
```

```python
# aggrigating cars and their types that do not qualify for clean alternative fuel vehicle using groupby function
car_Model=(data[data['Clean Alternative Fuel Vehicle (CAFV) Eligibility']!= 'Clean Alternative Fuel Vehicle Eligible'].groupby(['Make','Model']).sum()['Base MSRP'])
```

```python
# this arranges the make and model in dictionary format show tham in a presentable manner
car_and_model={}
k1='AUDI'
values=[]
for k,v in dict(car_Model).keys():
    if k == k1:
        values.append(v)
    else:
        car_and_model[k1]=values
        k1=k
        values=[]
print('cars and their types that do not qualify for clean alternative fuel vehicle are :- \n')
for k,v in car_and_model.items():
    print(f'{k}         {v}')
```

## Draw plots to show the distribution between city and electric vehicle type

```python
data
```

```python

# Filter the DataFrame to include only relevant columns
df = data[['City', 'Electric Vehicle Type']]

# Group the data by City and Electric Vehicle Type and calculate the count
grouped_data = df.groupby(['City', 'Electric Vehicle Type']).size().reset_index(name='Count')

# Pivot the data to create a matrix of City vs. Electric Vehicle Type counts
pivot_data = grouped_data.pivot(index='City', columns='Electric Vehicle Type', values='Count')

# Plot a stacked bar chart to show the distribution
pivot_data.plot(kind='bar', stacked=True, figsize=(12, 8))
plt.title('Distribution of Electric Vehicle Types by City')
plt.xlabel('City')
plt.ylabel('Count')
plt.legend(title='Electric Vehicle Type')
plt.xticks(rotation=45)
plt.show()
```

### the distribution is messy but we can conclude that fewer cities has larger concetration of cars lets check the distribution of car vs country

```python

# Filter the DataFrame to include only relevant columns
df = data[['County', 'Electric Vehicle Type']]

# Group the data by County and Electric Vehicle Type and calculate the count
grouped_data = df.groupby(['County', 'Electric Vehicle Type']).size().reset_index(name='Count')

# Pivot the data to create a matrix of County vs. Electric Vehicle Type counts
pivot_data = grouped_data.pivot(index='County', columns='Electric Vehicle Type', values='Count')

# Plot a stacked bar chart to show the distribution
pivot_data.plot(kind='bar', stacked=True, figsize=(12, 8))
plt.title('Distribution of Electric Vehicle Types by County')
plt.xlabel('County')
plt.ylabel('Count')
plt.legend(title='Electric Vehicle Type')
plt.xticks(rotation=45, ha='right')
plt.show()
```

### this distribution is like previous distribution fewer countries has larger concentration of cars
